```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
for i in data.columns:
   if data[i].dtype == 'object':
     data[i] = l.fit_transform(data[i])
```

**Dividing the dataset in two parts for training and testing**

```python
X = data.drop('Mental_fitness', axis=1)
Y = data['Mental_fitness']

from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.2, random_state=2)
```

**Applying different algorithms**

**1. Linear regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
# model training
lr = LinearRegression()
lr.fit(Xtrain, Ytrain)

# model evaluation on trained data
Ytrain_prediction = lr.predict(Xtrain)
mse = mean_squared_error(Ytrain, Ytrain_prediction)
rmse = (np.sqrt(mean_squared_error(Ytrain, Ytrain_prediction)))
r2 = r2_score(Ytrain, Ytrain_prediction)

print("The training performance of model")
print("---------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
# model evaluation on testing data
Ytest_prediction = lr.predict(Xtest)
mse = mean_squared_error(Ytest, Ytest_prediction)
rmse = (np.sqrt(mean_squared_error(Ytest, Ytest_prediction)))
r2 = r2_score(Ytest, Ytest_prediction)

print("The testing performance of model")
print("---------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

# **2. Random Forest Regression**

```python
from sklearn.ensemble import RandomForestRegressor
# model traning
rf = RandomForestRegressor()
rf.fit(Xtrain, Ytrain)

# model evaluating on trained data
Ytrain_prediction = rf.predict(Xtrain)
mse = mean_squared_error(Ytrain, Ytrain_prediction)
rmse = (np.sqrt(mean_squared_error(Ytrain, Ytrain_prediction)))
r2 = r2_score(Ytrain, Ytrain_prediction)

print("The training performance of model")
print("---------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
# Testing model and evaluation
Ytest_prediction = rf.predict(Xtest)
mse = mean_squared_error(Ytest, Ytest_prediction)
rmse = (np.sqrt(mean_squared_error(Ytest, Ytest_prediction)))
r2 = r2_score(Ytest, Ytest_prediction)

print("The testing performance of model")
print("---------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

**3. Decision Tree Regression**

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor(random_state=0)

dtr.fit(Xtrain, Ytrain)

# model evaluating on trained data
Ytrain_prediction = dtr.predict(Xtrain)
mse = mean_squared_error(Ytrain, Ytrain_prediction)
rmse = (np.sqrt(mean_squared_error(Ytrain, Ytrain_prediction)))
r2 = r2_score(Ytrain, Ytrain_prediction)

print("The training performance of model")
print("---------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
# model evaluating on Test data
Ytest_prediction = dtr.predict(Xtest)
mse = mean_squared_error(Ytest, Ytest_prediction)
rmse = (np.sqrt(mean_squared_error(Ytest, Ytest_prediction)))
r2 = r2_score(Ytest, Ytest_prediction)

print("The testing performance of model")
print("---------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

**4. SVM Regression**

```python
from sklearn.svm import SVR
svr = SVR()

svr.fit(Xtrain, Ytrain)

# model evaluating on trained data
Ytrain_prediction = svr.predict(Xtrain)
mse = mean_squared_error(Ytrain, Ytrain_prediction)
rmse = (np.sqrt(mean_squared_error(Ytrain, Ytrain_prediction)))
r2 = r2_score(Ytrain, Ytrain_prediction)

print("The training performance of model")
print("---------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
# model evaluating on Test data
Ytest_prediction = svr.predict(Xtest)
mse = mean_squared_error(Ytest, Ytest_prediction)
rmse = (np.sqrt(mean_squared_error(Ytest, Ytest_prediction)))
r2 = r2_score(Ytest, Ytest_prediction)

print("The testing performance of model")
print("---------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```
